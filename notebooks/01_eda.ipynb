```python
def melt_weather_data(df, var_name):
    return df.melt(id_vars=["datetime"], var_name="city", value_name=var_name)
```

```python
from functools import reduce

df_temp = melt_weather_data(temperature, "temperature")
df_humidity = melt_weather_data(humidity, "humidity")
df_pressure = melt_weather_data(pressure, "pressure")
df_wind_speed = melt_weather_data(wind_speed, "wind_speed")
df_wind_dir = melt_weather_data(wind_direction, "wind_direction")
df_weather_desc = melt_weather_data(weather_description, "weather_description")

# Merge all weather dataframes
dfs = [df_temp, df_humidity, df_pressure, df_wind_speed, df_wind_dir, df_weather_desc]
weather_data = reduce(lambda left, right: pd.merge(left, right, on=["datetime", "city"], how="outer"), dfs)

# Merge with city attributes
final_df = pd.merge(weather_data, city_attributes, left_on="city", right_on="City", how="left").drop(columns=["City"])

# Convert datetime and extract features
final_df["datetime"] = pd.to_datetime(final_df["datetime"], errors='coerce')
final_df["hour"] = final_df["datetime"].dt.hour
final_df["day"] = final_df["datetime"].dt.day
final_df["month"] = final_df["datetime"].dt.month
final_df["weekday"] = final_df["datetime"].dt.weekday

# Drop rows with missing datetime or city
final_df = final_df.dropna(subset=["datetime", "city"])
final_df
```

```python
df=final_df[['datetime', 'hour', 'day', 'month', 'weekday', 'city', 'Latitude',
       'Longitude', 'temperature', 'humidity', 'pressure', 'wind_speed',
       'wind_direction', 'weather_description']]
```

```python
# Visualizations
plt.figure(figsize=(12, 6))
sns.histplot(df["temperature"].dropna(), kde=True, bins=50)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (Kelvin)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=final_df, x="month", y="temperature")
plt.title("Monthly Temperature Distribution")
plt.xlabel("Month")
plt.ylabel("Temperature (Kelvin)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.scatterplot(data=final_df, x="humidity", y="temperature", alpha=0.3)
plt.title("Humidity vs. Temperature")
plt.xlabel("Humidity (%)")
plt.ylabel("Temperature (Kelvin)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
encoded_df = pd.get_dummies(df, columns=["weather_description"], dummy_na=True)
plt.figure(figsize=(14, 6))
encoded_df.filter(like="weather_description_").mean().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Weather Descriptions by Frequency")
plt.ylabel("Proportion of Records")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
